# Define PL Inputs

```python
# AND gate inputs
and_in_0 = GPIO(GPIO.get_gpio_pin(0), "out")
and_in_1 = GPIO(GPIO.get_gpio_pin(1), "out")

# OR gate inputs
or_in_0  = GPIO(GPIO.get_gpio_pin(2), "out")
or_in_1  = GPIO(GPIO.get_gpio_pin(3), "out")

# XOR gate inputs
xor_in_0 = GPIO(GPIO.get_gpio_pin(4), "out")
xor_in_1 = GPIO(GPIO.get_gpio_pin(5), "out")

# NOT gate inputs
not_in_0 = GPIO(GPIO.get_gpio_pin(6), "out")
```

# Define PL Outputs

```python
and_out = GPIO(GPIO.get_gpio_pin(16), "in")
or_out  = GPIO(GPIO.get_gpio_pin(17), "in")
xor_out = GPIO(GPIO.get_gpio_pin(18), "in")
not_out = GPIO(GPIO.get_gpio_pin(19), "in")
```

# Test gates

```python
print("AND")
print("| A | B | out |")
print("+---+---+-----+")
for a in [0, 1]:
    for b in [0, 1]:
        and_in_0.write(a)
        and_in_1.write(b)
        res = and_out.read()
        print(f"| {a} | {b} |  {res}  |")
```

```python
print("OR")
print("| A | B | out |")
print("+---+---+-----+")
for a in [0, 1]:
    for b in [0, 1]:
        or_in_0.write(a)
        or_in_1.write(b)
        res = or_out.read()
        print(f"| {a} | {b} |  {res}  |")
```

```python
print("XOR")
print("| A | B | out |")
print("+---+---+-----+")
for a in [0, 1]:
    for b in [0, 1]:
        xor_in_0.write(a)
        xor_in_1.write(b)
        res = xor_out.read()
        print(f"| {a} | {b} |  {res}  |")
```

```python
print("NOT")
print("| A | out |")
print("+---+-----+")
for a in [0, 1]:
    not_in_0.write(a)
    res = not_out.read()
    print(f"| {a} |  {res}  |")
```
